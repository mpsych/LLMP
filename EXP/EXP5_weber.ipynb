```python
def resultAnalysis(answer, label):
    import re
    match = re.search(r'(\d+)\s+extra\s+dots', answer)
    if match:
        result = int(match.group(1))
        error = result - label
        print("Result: ", result)
        print("GT:", label)
        print("Error: ", error)
    else:
        print("Invalid Answer")
```

Weber Experiment (10)

- Run1

```python
image, label = L.GPImage.weber('10')
imshow(image, cmap='gray')

question = 'What do you see? If you see dots, then proceed as follows: I initially added 10 black dots, then added extra dots to it (up to 10).Estimate the number of added dots in the image. It does not have to be precise; JUST give me your estimation (of the form "X extra dots")'
answer = L.ChatGPT.query(question, image)
resultAnalysis(answer, label)
```
