```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
    
import sys
sys.path.append('../')
import LLMP as L

from matplotlib.pyplot import imshow
```

```python
image, label = L.GPImage.figure1('angle')
imshow(image, cmap='gray')

#question = 'What do you see? If you see an angle, can you estimate it? Please respond with just the range'
#question = 'can you see an angle? can you estimate the angle?Please respond with just the range'
question = 'What do you see? if you have two lines how do you estimate the angle between them? based on what you said can you estimate angle between them? Please respond short answer with just the range'
answer = L.ChatGPT.query(question, image)
print(answer, 'GT:', label)
```

```python
import re
range = re.findall(r'\d+', answer)
range = [int(r) for r in range]
```

```python
range
```

```python
error_range = [label-r for r in range]
error_range
```
