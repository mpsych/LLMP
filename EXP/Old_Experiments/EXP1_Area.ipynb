```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
    
import sys
sys.path.append('../')
import LLMP as L

from matplotlib.pyplot import imshow
```

Run #1

```python
image, label = L.GPImage.figure1('area')
imshow(image, cmap='gray')

question = 'What do you see? If you see a circle, can you estimate the size of radius in pixels? if you can estimate the radius can you esitimate the area of the circle in pixels? Please respond with just the range'
answer = L.ChatGPT.query(question, image)
print(answer, 'GT:', label)
```

```python
import re
range = re.findall(r'\d+', answer)
range = [int(r) for r in range]
```

```python
range
```

```python
error_range = [label-r for r in range]
error_range
```
